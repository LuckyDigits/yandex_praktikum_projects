# Исследование пользователей мобильного приложения

В нашем распоряжении данные сервиса «Ненужные вещи». В нем пользователи продают свои ненужные вещи, размещая их на доске объявлений. Исходные данные содержат информацию о событиях, совершенных в мобильном приложении
сервиса, и источниках установки приложения.

Цель исследования — выделить сегменты пользователей на основе их поведения и определить стратегии по улучшению их вовлечённости.

Для реализации поставленной цели проанализируем поведение пользователей по следующим метрикам:
- частота совершения событий;
- время, проведенное в приложении;
- retention rate;
- конверсия в целевое действие. 

Так как данных о совершении сделок в приложении у нас нет, а с авторами объявлений можно связаться и вне приложения, то в качестве целевого действия в нашем случае выбран просмотр контактов в объявлении. 

Сегментировать пользователей будем по медианной продолжительности их сессий — одному из основных показателей для оценки вовлечённости пользователей. Для каждого сегмента подсчитаем и проанализируем retantion rate, конверсию. В завершении исследования выполним проверку статистических гипотез.

Описание данных:
1) mobile_sources.csv:
   - userId — идентификатор пользователя;
   - source — источник, с которого пользователь установил приложение.
2) mobile_dataset.csv:
   - event.time — время совершения;
   - user.id — идентификатор пользователя;
   - event.name — действие пользователя.

Виды действий event.name:
- advert_open — открыл карточки объявления;
- photos_show — просмотрел фотографий в объявлении;
- tips_show — увидел рекомендованные объявления;
- tips_click — кликнул по рекомендованному объявлению;
- contacts_show и show_contacts — посмотрел номер телефона;
- contacts_call — позвонил по номеру из объявления;
- map — открыл карту объявлений;
- search_1 — search_7 — разные действия, связанные с поиском по сайт;
- favorites_add — добавил объявление в избранное.

### Время в приложении

Подсчитаем время, проведенное пользователями в приложении. Для этого, определим пользовательские сессии — временной интервал, в течение которого пользователь взаимодействовал с сервисом. Сессия заканчивается, когда в течение определённого времени пользователь не совершает новых действий. В нашем расчёте сессий будем считать, что новая сессия начинается после 30 минут бездействия (такой же интервал при расчёте сессий используют, например, Google Analytics и Яндекс.Метрика). Также, добавим столбец с временным интервалом между событиями, это поможет нам подсчитать длительность каждой сессии.

```python
# сортируем данные
df = mobile_dataset.sort_values(['user_id', 'event_time'])

# определяем границы сессии
g = (df.groupby('user_id')['event_time'].diff() > pd.Timedelta(minutes=30)).cumsum()

# считаем сессии
df['session_id'] = df.groupby(['user_id', g], sort=False).ngroup() + 1

# добавляем столбец с длительностью
df['duration'] = df.groupby('session_id')['event_time'].diff().dt.total_seconds()
```

```python
df.head(10)
```

Данные подготовлены, посмотрим сколько времени суммарно пользователи провели в приложении в каждый из дней.

```python
time_in_app_by_date = df.groupby('dt')['duration'].sum().reset_index()
sns.set(rc={"figure.figsize":(16 , 10)})
sns.barplot(x='dt', y='duration', data=time_in_app_by_date, color='steelblue', ci= None)
plt.xticks(rotation=90)
plt.xlabel('Дата')
plt.ylabel('Суммарное время, с')
plt.title('Суммарное время, проведённое пользователями в приложении')
```

Суммарная продолжительность распределена по дням неравномерно. Максимальное количество времени в приложении пользователи провели 29 октября, минимальное — 12 октября.

Составим данные о пользовательских сессиях в отдельный датасет.

```python
sessions = df.groupby(['user_id', 'session_id']).agg({'event_name': 'count', 'event_time': ['min', 'max']}).reset_index()
sessions.columns = ['user_id', 'session_id', 'events_cnt', 'session_start', 'session_end']
sessions['duration'] = round((sessions['session_end'] - sessions['session_start']).dt.total_seconds())
sessions.head(10)
```

```python
sessions.plot(kind='hist', y='duration', bins=100, linewidth=1, alpha=1, grid=True, legend=True,
              title='Распределение сессий по длительности', figsize=(12, 6)) 
plt.xlabel("Длительность, с")
plt.ylabel("Количество сессий")
```

По графику большая часть сессий длилась примерно до 60 минут. Подсчитаем перцентили.

```python
np.percentile((sessions['duration'] / 60).round(2), [1, 5, 10, 25, 50, 75, 90, 95, 99])
```

Действительно, 95% сессий длились менее 50 минут.

Очистим данные от аномалий.

```python
shape = df.shape[0] # сохраним размер датасета до удаления аномальных сессий
```

```python
# сохраняем очищеные данные
sessions_elimination1 = list(sessions.query('duration > 7200')['session_id'])
sessions_elimination2 = list(sessions.query('duration == 0 and events_cnt > 1')['session_id'])
sessions_elimination3 = list(df.query('session_id in @zero_duration and event_name not in ["tips_show", "advert_open"]')['session_id'])
sessions_clean = sessions.query('session_id not in @sessions_elimination1 and session_id not in @sessions_elimination2 and session_id not in @sessions_elimination3')
df = df.query('session_id not in @sessions_elimination1 and session_id not in @sessions_elimination2 and session_id not in @sessions_elimination3')
```

```python
print(f'Удалено событий: {shape - len(df)} ({round((1 - len(df) / shape) * 100, 2)}% событий)')
print(f'Удалено сессий: {len(sessions) - len(sessions_clean)} ({round((1 - (len(sessions_clean) / len(sessions))) * 100, 1)}% сессий)')
```

После удаления аномалий было потеряно около 12% от всех сессий и около 5,2% от всех событий.

В очищенных данных переведём продолжительность сессий в минуты.

```python
sessions_clean['duration'] = (sessions_clean['duration'] / 60).round(2)
```

```python
np.percentile(sessions_clean['duration'].round(1), [1, 5, 10, 25, 50, 60, 75, 90, 95, 99])
```

Медианная продолжительность сессии составляет — 7.4 минуты, средняя — около 14 минут.

### Retention Rate

Составим профили пользователей: сохраним информацию о первом посещении приложения для каждого пользователя.

Рассчитаем Retention Rate.

```python
def get_retention(
    profiles,
    sessions,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')

    # собираем «сырые» данные для расчёта удержания
    result_raw = result_raw.merge(
        sessions[['user_id', 'session_start']], on='user_id', how='left'
    )
    result_raw['lifetime'] = (
        result_raw['session_start'] - result_raw['first_ts']
    ).dt.days

    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users' 
        dimensions = dimensions + ['cohort']
        
    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        result = result.div(result['cohort_size'], axis=0)
        result = result[['cohort_size'] + list(range(horizon_days))]
        result['cohort_size'] = cohort_sizes
        return result

    # получаем таблицу удержания
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)

    # для таблицы динамики удержания убираем 'cohort' из dimensions
    if 'cohort' in dimensions: 
        dimensions = []

    # получаем таблицу динамики удержания
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )

    # возвращаем обе таблицы и сырые данные
    return result_raw, result_grouped, result_in_time
```

Расчитаем конверсию в целевое действие при помощи когортного анализа. Посмотрим общую динамику по всем пользователям.

```python
def get_conversion(
    profiles,
    data,
    observation_date,
    horizon_days,
    dimensions=[],
    ignore_horizon=False,
):

    # исключаем пользователей, не «доживших» до горизонта анализа
    last_suitable_acquisition_date = observation_date
    if not ignore_horizon:
        last_suitable_acquisition_date = observation_date - timedelta(
            days=horizon_days - 1
        )
    result_raw = profiles.query('dt <= @last_suitable_acquisition_date')

    # определяем дату и время первого целевого действия для каждого пользователя
    first_show_contacts = (
        data.query('event_name == "show_contacts"').sort_values(by=['user_id', 'dt'])
        .groupby('user_id')
        .agg({'event_time': 'first'})
        .rename(columns={'event_time': 'event_dt'})
        .reset_index()
    )

    # добавляем данные о целевых действиях в профили
    result_raw = result_raw.merge(
        first_show_contacts[['user_id', 'event_dt']], on='user_id', how='left'
    )

    # рассчитываем лайфтайм 
    result_raw['lifetime'] = (
        result_raw['event_dt'] - result_raw['first_ts']
    ).dt.days

    # группируем по cohort, если в dimensions ничего нет
    if len(dimensions) == 0:
        result_raw['cohort'] = 'All users' 
        dimensions = dimensions + ['cohort']

    # функция для группировки таблицы по желаемым признакам
    def group_by_dimensions(df, dims, horizon_days):
        result = df.pivot_table(
            index=dims, columns='lifetime', values='user_id', aggfunc='nunique'
        )
        result = result.fillna(0).cumsum(axis = 1)
        cohort_sizes = (
            df.groupby(dims)
            .agg({'user_id': 'nunique'})
            .rename(columns={'user_id': 'cohort_size'})
        )
        result = cohort_sizes.merge(result, on=dims, how='left').fillna(0)
        # делим каждую «ячейку» в строке на размер когорты
        # и получаем conversion rate
        result = result.div(result['cohort_size'], axis=0)
        result = result[['cohort_size'] + list(range(horizon_days))]
        result['cohort_size'] = cohort_sizes
        return result

    # получаем таблицу конверсии
    result_grouped = group_by_dimensions(result_raw, dimensions, horizon_days)

    # для таблицы динамики конверсии убираем 'cohort' из dimensions
    if 'cohort' in dimensions: 
        dimensions = []

    # получаем таблицу динамики конверсии
    result_in_time = group_by_dimensions(
        result_raw, dimensions + ['dt'], horizon_days
    )

    # возвращаем обе таблицы и сырые данные
    return result_raw, result_grouped, result_in_time
```

Разделим пользователей на 3 сегмента в зависимости от медианной продолжительности их сессий:
1) `"short"` — с короткими сессиями;
2) `"median"` — с средними сессиями;
3) `"long"` — с длинными сессиями.

За короткие сессии примем сессии с продолжительностью менее 2,6 минуты (квантиль 0,3). За длинные — значения более 11,15 минуты (квантиль 0,6). ## Статистический анализ

### Гипотеза 1

Гипотеза — пользователи из разных источников показывают разную конверсию в просмотры контактов.

```python
# добавляем информацию об источниках установки
df = df.merge(mobile_sourсes, on='user_id', how='left')
```

Для начала посмотрим как пользователи распределены по источникам установки приложения.

```python
df.groupby('source').count().plot(kind='pie', y='user_id', autopct='%1.0f%%',
                                  title='Соотношение пользователей по источнику установки', legend=False,
                                  figsize=(12, 6))
```

Больше всего пользователей установили приложение через Yandex. Следом за ними по числу установок — Google.

Проверим гипотезу на пользователях, установивших приложение через Google или Yandex:
- Нулевая гипотеза — различий в долях пользователей, совершивших целевое действие, между группами нет;
- Альтернативная — различия между группами есть.

```python
alpha = 0.05 # уровень статистической значимости
success1 = df.query('source == "google" and event_name == "show_contacts"')['user_id'].nunique()
success2 = df.query('source == "yandex" and event_name == "show_contacts"')['user_id'].nunique()
trial1 = df.query('source == "google"')['user_id'].nunique()
trial2 = df.query('source == "yandex"')['user_id'].nunique()
p1 = success1 / trial1
p2 = success2 / trial2
p_combined = (success1 + success2) / (trial1 + trial2)
difference = p1 - p2 
z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trial1 + 1/trial2))
distr = stats.norm(0, 1)  
p_value = (1 - distr.cdf(abs(z_value))) * 2
print(f'Сравнение групп "Google" - "Yandex"')
print('P-value:', '{0:.3f}'.format(p_value))
if p_value < alpha:
    print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
else:
    print('Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными')
```

P-value больше критического уровня статистической значимости — различия в долях пользователей, совершивших целевое действие, между группами статистически незначительны.

- Альтернативная — различия между группами есть.

```python
# добавляем информацию о сегментах
df = df.merge(profiles[['user_id', 'category']], on='user_id', how='left')
```

```python
alpha = 0.05
success1 = df.query('category == "short" and event_name == "show_contacts"')['user_id'].nunique()
success2 = df.query('category == "long" and event_name == "show_contacts"')['user_id'].nunique()
trial1 = df.query('category == "short"')['user_id'].nunique()
trial2 = df.query('category == "long"')['user_id'].nunique()
p1 = success1 / trial1
p2 = success2 / trial2
p_combined = (success1 + success2) / (trial1 + trial2)
difference = p1 - p2 
z_value = difference / mth.sqrt(p_combined * (1 - p_combined) * (1/trial1 + 1/trial2))
distr = stats.norm(0, 1)  
p_value = (1 - distr.cdf(abs(z_value))) * 2
print(f'Сравнение групп "short" - "long"')
print('P-value:', '{0:.3f}'.format(p_value))
if p_value < alpha:
    print('Отвергаем нулевую гипотезу: между долями есть значимая разница')
else:
    print('Не получилось отвергнуть нулевую гипотезу, нет оснований считать доли разными')
```

P-value больше критического уровня статистической значимости — различия в долях пользователей, совершивших целевое действие, между группами статистически незначительны. 

### Вывод

Статистически значимых различий в конверсиях между пользователями, установившими приложение через Google или Yandex по количеству пользователей — похоже, источник установки не оказывает большое влияние на конверсию.

Конверсия пользователей с короткой медианной продолжительностью сессии выше, чем у пользователей с длинной медианной продолжительностью сессии, но статистически значимой разницы между ними не обнаружено. Принимая во внимание расчёты конверсии в предыдущем пункте, отсутствие статистически значимой разницы между группами можно объяснить тем, что конверсия пользователей с короткими сессиями растёт более резко по сравнению с конверсией пользователей с длинными сессиями. Таким образом, за счёт низкого старта в сочетании с резким ростом она достигает того же соотношения пользователей, совершивших целевое действие, что и сегмент пользователей с длинными сессиями (у которого конверсия изначально более высокая, но в течении первых двух недель практически не растёт).
