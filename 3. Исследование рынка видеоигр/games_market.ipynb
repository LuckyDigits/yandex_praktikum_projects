# Анализ рынка видеоигр

В нашем распоряжении данные из открытых источников о продажах игр, оценках пользователей и экспертов, жанрах и платформах (например, Xbox или PlayStation). Временные рамки — до 2016 года. Цель исследования — определить закономерности, влияющие на успешность игры. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании. 

**Описание данных:**
* Name — название игры;
* Platform — платформа;
* Year_of_Release — год выпуска;
* Genre — жанр игры;
* NA_sales — продажи в Северной Америке (миллионы проданных копий);
* EU_sales — продажи в Европе (миллионы проданных копий);
* JP_sales — продажи в Японии (миллионы проданных копий);
* Other_sales — продажи в других странах (миллионы проданных копий);
* Critic_Score — оценка критиков (максимум 100);
* User_Score — оценка пользователей (максимум 10);
* Rating — рейтинг от организации ESRB (англ. Entertainment Software Rating Board). Эта ассоциация определяет рейтинг компьютерных игр и присваивает им подходящую возрастную категорию.

## Предобработка данных

### Изменение названий столбцов 

Приведём названия столбцов к нижнему регистру.

```python
data.columns = data.columns.str.lower()
```

```python
data.head(15)
```

### Обработка пропусков

```python
data.isna().sum()
```

Посмотрим на пропуски в столбце `name` — какие ещё данные отсутствуют в этих двух строках.

```python
data[data['name'].isna()]
```

Данные здесь отсутствуют одновременно в 5 столбцах. Смысла оставлять эти строки нет. Также, избавимся от пропусков в столбце `year_of_release` — их в датасете 269 строк. При таком количестве их удаление не должно существенно повлиять на последующий анализ.

```python
data = data.dropna(subset=['name', 'year_of_release']).reset_index(drop=True)
```

```python
data.isna().sum()
```

Пропуски в столбцах `critic_score`, `user_score`, `rating` оставим как есть, так как оценки и рейтинг отличаются от платформы к платформе.

### Обработка дубликатов

```python
data.duplicated().sum()
```

Явных дубликатов нет.

Посмотрим на уникальные значения присутствуют в столбце `user_score`.

```python
data['user_score'].unique()
```

```python
len(data[data['user_score'] == 'tbd'])
```

TBD — аббревиатура от английского To Be Determined (будет определено) или To Be Decided (будет решено). Она используется, если какая-то информация еще не определена или решение по вопросу не принято. Заменим `tbd` на `NaN`, чтобы не плодить лишние сущности.

```python
data.loc[data['user_score'] == 'tbd', 'user_score'] = float('NaN')
```

Рассмотрим уникальные значения в столбце `rating`.

```python
data['rating'].unique()
```

"К-A" — это базовый рейтинг. Игры с таким рейтингом имеют контент, который "в целом подходит для всех возрастов". Они могут содержать незначительные случаи мультипликационного насилия или комического вреда. После 1998 года этот рейтинг получил название "E".

Устраним эти дубликаты.

```python
data['rating'] = data['rating'].replace('K-A', 'E')
```

Также, проверим, если ли повторяющиеся строки в столбцах `name`, `platform`, `year_of_release`. Удалим их, если они есть.

```python
duplicates = data[data.duplicated(['name', 'platform', 'year_of_release'])]
duplicates
```

```python
data = data.drop_duplicates(subset=['name', 'platform', 'year_of_release'])
```

### Изменение типов данных 

Для удобства преобразуем данные в `year_of_release` в тип `int`.

```python
data['year_of_release'] = data['year_of_release'].astype('int')
```

А данные в `user_score` из `object` преобразуем в `float`.

```python
data['user_score'] = data['user_score'].astype('float')
```

### Подсчёт суммарных продаж

Добавим столбец `sales_total` с суммарными продажами игр по всем регионам.

```python
data['sales_total'] = data['na_sales'] + data['eu_sales'] + data['jp_sales'] + data['other_sales']
```

Посмотрим что получилось.

```python
data.head(10)
```

### Продажи по платформам

Определим платформы с наибольшими суммарными продажами. Для этого, сначала агрегируем наши суммарные продажи по платформам — т.е. подсчитаем сколько всего было продаж на той или иной платформе за всё время наблюдений. После, оставим только те платформы, у которых суммарные продажи превышают 100 миллионов проданных копий. Визуализируем данные через тепловую карту.

```python
data_temp1 = data.pivot_table(index='platform', values='sales_total', aggfunc='sum')
data_temp1 = data_temp1.reset_index()
data_temp1.columns = ['platform', 'sales_total']
top_by_sales = data_temp1.sort_values(by='sales_total', ascending=False).query('sales_total > 100')
```

```python
top_plaforms = top_by_sales['platform'].tolist()
data_temp2 = data.query('platform == @top_plaforms')
data_pivot = data_temp2.pivot_table(index='platform', columns='year_of_release', values='sales_total', aggfunc='sum', fill_value=0).round()
plt.figure(figsize=(20, 10))
sns.heatmap(data_pivot, annot = True, fmt='.3g', cmap='PuBu', robust=True, linecolor='black', linewidths=.1)
plt.xlabel('Год')
plt.ylabel('Платформа')
plt.xticks(rotation=90) 
plt.title('Изменение суммарных продаж по платформам за всё время наблюдений')
```

```python
lifetime = []
for p in data['platform'].unique():
    platform_years = data.query('platform == @p')['year_of_release']
    platform_lifetime = platform_years.max() - platform_years.min() + 1
    lifetime.append(platform_lifetime)
    if platform_lifetime == 1:
        print(f'Срок жизни платформы {p} — {platform_lifetime} год')
    elif platform_lifetime >= 2 and platform_lifetime <= 4:
        print(f'Срок жизни платформы {p} — {platform_lifetime} года')
    elif platform_lifetime >= 5:
        print(f'Срок жизни платформы {p} — {platform_lifetime} лет') 
```

```python
print(f'Средий срок жизни платформы — {statistics.median(lifetime)} лет.')
```

На тепловой карте видно, что за всё время наблюдений наибольшее количество продаж было совершено на платформах: PS2, X360, PS3, Wii, DS, PS. Но, это не говорит о том, что они обязательно являются потенциально прибыльными, так как на некоторых из них к 2016 году уже перестали выходить новые игры.

### Продажи на популярных платформах за актуальный период

В качестве актуального выберем период с 2014 по 2016 годы. Период до 2014 года, как мне кажется, включать в анализ для определения потенциально наиболее прибыльных в будущем платформ не стоит, так как в него могут попасть уже неактуальные платформы.

```python
data_actual = data.query('year_of_release > 2013')
data_actual = data_actual.pivot_table(index=['platform',
                                             'year_of_release'], values='sales_total', aggfunc='sum', fill_value=0)
data_actual = data_actual.reset_index()
sns.set(rc={"figure.figsize":(16, 8)})
sns.barplot(x='year_of_release', y='sales_total',
            hue='platform', data=data_actual.sort_values(by='sales_total', ascending=False), ci= None)
plt.legend(loc='upper right')
plt.xlabel('Год')
plt.ylabel('Суммарные продажи')
plt.title('Изменение суммарных продаж по платформам за актуальный период')
```

Наибольшие суммарные продажи за актуальный период имеют платформы — PS4, XOne и 3DS. В целом, актуальными платформами можно считать PS4, XOne, 3DS, PC, WiiU, PSV.

Построим диаграмму размаха по глобальным продажам игр в разбивке по платформам. График будем строить по данным за актуальный период.

```python
boxplot = sns.boxplot(x="platform", y="sales_total", data=data.query('year_of_release > 2013'))
plt.figure(figsize=(12, 6))
boxplot.axes.set_title("Диаграммы размаха потенциально прибыльных платформ")
boxplot.set_xlabel("Платформа")
boxplot.set_ylabel("Суммарные продажи")
boxplot.set_ylim(0, 1.6)
```

Выбросы здесь — это игры, принёсшие неожиданно много денег. Наибольшее количество таких игр у платформ — PS4, 3DS, XOne.

Большое количество выбросов при небольшом размахе как раз говорит нам о том, что такие неожиданно "выстрелившие" игры приносят данным платформам основную прибыль. В этом контексте интересны платформы PSV, PC и 3DS. 

В целом, диаграмма размаха подтверждает правильность ранее выбранных платформ в качестве актуальных, так как мы видим, что:
* PSP кажется не приносит почти никаких денег;
* Wii в целом имеет неплохой размах, но, учитывая что на рынке уже не первый год находится её последовательница WiiU, вряд ли имеет смысл инвестировать в выход игр на более старой платформе (то же самое касается PS3 и X360).

### Оценка влияния отзывов пользователей и критиков на продажи

Для начала, оценим влияние отзывов пользователей и критиков на примере PS3 — в рассматриваемый нами период она всё ещё приносила деньги, даже несмотря на то, что уже сильно уступала PS4.

Построим диаграмму рассеяния и посчитаем корреляцию между отзывами и продажами.

```python
ps3 = data.query('platform == "PS3" and year_of_release > 2013')
ps3.plot(kind='scatter', x='critic_score', y='sales_total', figsize=(16, 8))
plt.xlabel('Оценки критиков')
plt.ylabel('Суммарные продажи')
plt.title('Зависимость между оценками критиков и суммарными продажами на платформе PS3')
```

```python
print('Коэффициент корреляции Пирсона между оценкой критиков и суммарными продажами:', (ps3['critic_score'].corr(ps3['sales_total'])).round(3))
```

Зависимость между оценками критиков и суммарными продажами слабая.

```python
ps3.plot(kind='scatter', x='user_score', y='sales_total', figsize=(16, 8))
plt.xlabel('Оценки пользователей')
plt.ylabel('Суммарные продажи')
plt.title('Зависимость между оценками пользователей и суммарными продажами на платформе PS3')
```

```python
print('Коэффициент корреляции Пирсона между оценкой пользователей и суммарными продажами:', (ps3['user_score'].corr(ps3['sales_total'])).round(3))
```

Зависимости между оценками пользователей и суммарными продажами нет или она не линейная.

Также, проверим, есть ли зависимость между оценками критиков и пользователей.

```python
ps3.plot(kind='scatter', x='user_score', y='critic_score', figsize=(16, 8))
plt.xlabel('Оценки пользователей')
plt.ylabel('Оценки критиков')
plt.title('Зависимость между оценками пользователей и критиков на платформе PS3')
```

```python
print('Коэффициент корреляции Пирсона между оценкой пользователей и критиков:', (ps3['user_score'].corr(ps3['critic_score'])).round(3))
```

А здесь зависимость прослеживается уже более отчетливо: с ростом оценок критиков растут и оценки пользователей.

Рассчитаем те же зависимости и для всех актуальных платформ.

```python
def platform_info_corr(platform):
    print('Платформа:', platform)
    data_platform = data.loc[(data['platform'] == platform) & (data['critic_score'] > 0) & (data['user_score'] > 0) & (data['year_of_release'] > 2013)]
    print('Коэффициент корреляции Пирсона между оценкой критиков и суммарными продажами:', data_platform['critic_score'].corr(data_platform['sales_total']).round(3))
    print('Коэффициент корреляции Пирсона между оценкой пользователей и суммарными продажами:', data_platform['user_score'].corr(data_platform['sales_total']).round(3))
    print('Коэффициент корреляции Пирсона между оценками пользователей и критиков:', data_platform['critic_score'].corr(data_platform['user_score']).round(3))
    pd.plotting.scatter_matrix(data_platform[['user_score', 'critic_score', 'sales_total']], figsize=(10, 10))
```

## Проверка гипотез

### Гипотеза 1

Гипотеза 1 — средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

* Нулевая гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые.
* Альтернативная гипотеза: средние пользовательские рейтинги платформ Xbox One и PC отличаются.

```python
data_xone = data.loc[(data['platform'] == 'XOne') & (data['user_score'].notnull()) & (data['year_of_release'] > 2013)]
data_pc = data.loc[(data['platform'] == 'PC') & (data['user_score'].notnull()) & (data['year_of_release'] > 2013)]
alpha = 0.05  # уровень статистической значимости
results = st.ttest_ind(data_xone['user_score'],
                       data_pc['user_score'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Получается, средние пользовательские рейтинги по этим платформам не отличаются статистически значимо между собой. Посмотрим, чему именно они равны.

```python
print('Средняя пользовательская оценка на платформе Xbox One —', data_xone['user_score'].mean().round(2))
print('Средняя пользовательская оценка на PC —', data_pc['user_score'].mean().round(2))
```

```python
data_XOne_PC = data.query('platform == ["XOne", "PC"] and year_of_release > 2013')
sns.set (rc={"figure.figsize":(12 , 6)})
sns.barplot(x='year_of_release', y='user_score', hue='platform', data=data_XOne_PC, ci= None)
plt.xlabel('Год')
plt.ylabel('Средняя оценка пользователей')
plt.title('Сравнение средних оценок пользователей платформ XOne и PC за актуальный период')
```

### Гипотеза 2

Гипотеза 2 — средние пользовательские рейтинги жанров Action и Sports  разные.

* Нулевая гипотеза: средние пользовательские рейтинги жанров Action и Sports одинаковые.
* Альтернативная гипотеза: средние пользовательские рейтинги жанров Action и Sports отличаются.

```python
data_action = data.loc[(data['genre'] == 'Action') & (data['user_score'].notnull()) & (data['year_of_release'] > 2013)]
data_sports = data.loc[(data['genre'] == 'Sports') & (data['user_score'].notnull()) & (data['year_of_release'] > 2013)]
alpha = 0.05  # уровень статистической значимости
results = st.ttest_ind(data_action['user_score'],
                       data_sports['user_score'])

print('p-значение:', results.pvalue)

if results.pvalue < alpha:
    print('Отвергаем нулевую гипотезу')
else:
    print('Не получилось отвергнуть нулевую гипотезу')
```

Нулевая гипотеза отвергнута, значит, средние пользовательские рейтинги двух жанров отличаются статистически значимо друг от друга. Посмотрим насколько.

```python
print('Средний пользовательский рейтинг игр в жанре Action —', data_action['user_score'].mean().round(2))
print('Средний пользовательский рейтинг игр в жанре Sports —', data_sports['user_score'].mean().round(2))
```

```python
data_action_sports = data.query('genre == ["Action", "Sports"] and year_of_release > 2013')
sns.set (rc={"figure.figsize":(12, 6)})
sns.barplot(x='year_of_release', y='user_score', hue='genre', data=data_action_sports, ci= None)
plt.xlabel('Год')
plt.ylabel('Средняя оценка пользователей')
plt.title('Сравнение средних оценок пользователей игр в жанрах Action и Sports за актуальный период')
```

**Вывод: средние пользовательские оценки на платформах Xbox One и PC не отличаются статистически значимо друг от друга. Средний пользовательский рейтинг игр в жанре Action статистически значимо выше чем у игр в жанре Sports.**
