# Исследование объявлений о продаже квартир

В нашем распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктов за  2014-2019 гг. Необходимо определить параметры, оказывающие наибольшее влияние на рыночную стоимость объектов недвижимости. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые — получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма. 

Описание данных:
- `airports_nearest` — расстояние до ближайшего аэропорта в метрах (м);
- `balcony` — число балконов;
- `ceiling_height` — высота потолков (м);
- `cityCenters_nearest` — расстояние до центра города (м);
- `days_exposition` — сколько дней было размещено объявление (от публикации до снятия);
- `first_day_exposition` — дата публикации;
- `floor` — этаж;
- `floors_total` — всего этажей в доме;
- `is_apartment` — апартаменты (булев тип);
- `kitchen_area` — площадь кухни в квадратных метрах (м²);
- `last_price` — цена на момент снятия с публикации;
- `living_area` — жилая площадь в квадратных метрах (м²);
- `locality_name` — название населённого пункта;
- `open_plan` — свободная планировка (булев тип);
- `parks_around3000` — число парков в радиусе 3 км;
- `parks_nearest` — расстояние до ближайшего парка (м);
- `ponds_around3000` — число водоёмов в радиусе 3 км;
- `ponds_nearest` — расстояние до ближайшего водоёма (м);
- `rooms` — число комнат;
- `studio` — квартира-студия (булев тип);
- `total_area` — общая площадь квартиры в квадратных метрах (м²);
- `total_images` — число фотографий квартиры в объявлении.

## Предобработка данных

###  Обработка пропусков

```python
data.isna().sum()
```

Заполним пропуски там, где это возможно.

#### Столбец `balcony`

Если продавец не указал количество балконов, значит можно сделать вывод, что в его квартире их нет. Поэтому, пропущенные значения в этом столбце заполним значением 0.

```python
data['balcony'] = data['balcony'].fillna(0) 
```

#### Столбцы `living_area`, `kitchen_area`

Заполним пропуски в столбце `living_area` с следующим образом: 
1) найдем соотношение медианной общей площади к медианной жилой;
2) заполним пропуски отношением общей площади к найденному соотношению.

С пропусками в столбце `kitchen_area` поступим аналогичным образом.

```python
#сохраним в переменной 'temp1' соотношение медианной общей площади к медианной жилой
temp1 = data['total_area'].median() / data['living_area'].median()
#заполним пропущенные данные в 'living_area' отношением общей площади к 'temp1'
data.loc[data['living_area'].isna() == True, 'living_area'] = data['total_area'] / temp1
```

```python
#сохраним в переменной 'temp2' соотношение медианной общей площади к медианной площади кухни
temp2 = data['total_area'].median() / data['kitchen_area'].median()
#заполним пропущенные данные в 'kitchen_area' отношением общей площади к 'temp2'
data.loc[data['kitchen_area'].isna() == True, 'kitchen_area'] = data['total_area'] / temp2
```

```python
data['living_area'].isna().sum(), data['kitchen_area'].isna().sum()
```

Пропуски заполнены.

#### Столбец `ceiling_height`

Гипотеза: высота потолков зависит от близости квартиры к центру города, особенно в Петербурге — чем ближе к центру города, тем  больше исторической застройки, а значит и выше потолки. Для проверки этой гипотезы категоризируем данные в столбце `cityCenters_nearest` по близости к центру города и посмотрим их медианную высоту потолков.

```python
#категоризирем данные 
def categorize_area(area):
    try:
        if 0 <= area <= 5000:
            return 'A'
        elif 5001 <= area <= 10000:
            return 'B'
        elif 10001 <= area <= 15000:
            return 'C'
        elif 15001 <= area <= 20000:
            return 'D'
        elif 20001 <= area <= 25000:
            return 'E'
        elif 25001 <= area <= 30000:
            return 'F'
        elif area >= 30001:
            return 'G'
    except:
        pass
```

```python
#применяем категоризацию к  столбцу 'cityCenters_nearest'
data['area_categorized'] = data['cityCenters_nearest'].apply(categorize_area)
```

```python
#постром сводную таблицу для проверки гипотезы
data_temp = data.pivot_table(index='area_categorized', values='ceiling_height', aggfunc=['count', 'median'])
data_temp.columns = ['Всего значений', 'Медианная высота']
data_temp
```

Получившаяся таблица в целом подтверждает гипотезу о зависимости высоты потолков от близости к центру города (данные по категориям E, F выбиваются из общей картины, что скорее всего можно объяснить меньшим количеством данных по ним).

Заполним пропуски медианными значениями по каждой категории.

```python
#заполним пропуски в 'ceiling_height' медианными значениями по каждой категории
for x in data['area_categorized'].unique():
    data.loc[(data['area_categorized'] == x) & (data['ceiling_height'].isna()), 'ceiling_height'] = \
    data.loc[(data['area_categorized'] == x), 'ceiling_height'].median()
```

Проверим, сколько пропусков осталось после заполнения.

```python
data['ceiling_height'].isna().sum()
```

Осталось 2575 пропусков — в этих строках нет данных о расположении квартиры относительно центра города, поэтому оставляем их как есть.

#### Столбец `floors_total`

В этом столбце 86 пропусков. Воспользуемся той же логикой, что и при заполнении пропусков в `ceiling_height` — высота здания зависит от его близости к центру города (обычно в центре города располагаются невысокие исторические здания, в спальных районах — многоэтажки).

Заполним, где возможно, пропуски медианными значениями по категориям.

```python
#заполним пропуски в 'floors_total' медианными значениями по каждой категории
for x in data['area_categorized'].unique():
    data.loc[(data['area_categorized'] == x) & (data['floors_total'].isna()), 'floors_total'] = \
    data.loc[(data['area_categorized'] == x), 'floors_total'].median()
```

Подсчитаем оставшиеся после заполнения пропуски.

```python
data['floors_total'].isna().sum()
```

```python
data[data['floors_total'].isna() == True]
```

Как мы видим, в этих строках также отсутствуют данные и во многих других столбцах, поэтому предлагаю их удалить.

```python
data = data.dropna(subset=['floors_total']) 
```

#### Cтолбец `locality_name`

В этом столбце пропусков совсем немного — 49 строк. Их удаление не должно повлиять на общую картину.

```python
data = data.dropna(subset=['locality_name'])
```

#### Возможные причины пропусков в данных

1. Столбцы ***`cityCenters_nearest`, `airports_nearest`, `parks_around3000`, `parks_nearest`, `ponds_around3000`, `ponds_nearest`***

Данные в них были получены автоматически из геосервисов — возможно имел место сбой при заполнении/выгрузке данных.

2. Столбцы ***`balcony`, `floors_total`, `ceiling_height`, `kitchen_area`, `living_area`***

Возможно, данные по этим параметрам были не обязательны к заполнению в объявлениях, поэтому пользователи их пропустили. Или же, пользователи могли указать их непосредственно в тексте объявлений, а не в форме.

3. Столбец ***`is_apartment`***

Пользователи могут не очень понимать разницу между апартаментами и просто квартирами и, поэтому, просто пропускать эту графу в своих объявлениях.

4. Столбец ***`days_exposition`*** 

Здесь пропуски скорее всего можно объяснить тем, что дата закрытия сделки не была указана пользователем — сделка была завершена вне сервиса.

#### Тип этажа квартиры

Категоризируем этажи квартир на следующие группы:
- первый;
- последний;
- другой.

```python
#создаем функцию для категоризации этажей на 'первый' и 'другой'
def categorize_level(floor):
    try:
        if floor == 1:
            return 'первый'
        else:
            return 'другой'
    except:
        pass
```

#### Общяя площадь

```python
def que(category, style):
    data_temp = data.pivot_table(index = category, values = 'last_price', aggfunc = ['count', 'median'])
    data_temp.columns = ['Количество', 'Медиана']
    data_temp.plot(y = 'Медиана', style=style, grid=True, figsize=(15, 10))
    print('Коэффициент корреляции Пирсона равен', (data['last_price'].corr(data[category])).round(3))
```

```python
que('total_area', 'o')
plt.title('Зависимость стоимости объектов от их общей площади')
plt.xlabel('Общая площадь, кв.м.')
plt.ylabel('Медианная стоимость')
plt.show()
```

Коэффициент корреляции Пирсона между общей площадью и стоимостью объекта равен 0,65. Связь между этими параметрами есть: в большинстве случаев чем больше площадь, тем выше стоимость.
